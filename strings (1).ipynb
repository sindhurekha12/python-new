```python
# anoymous function or unamed function

lambda arguments:expression
```

```python
def add(a,b):
    return a+b
add(10,20)
```

```python
add=lambda a,b:a+b
add(20,90)
```

```python
square=lambda a:a**2
square(10)
```

```python
#filter------->filter the records

#syntax: filter(function, iterable)

x=[23,4,5,17,89,90,7,34,256]

even=filter(lambda a:a%2==0,x)
list(even)
```

```python
names=["priya","sam","vicky","sindhu","krish"]

filtered_names=filter(lambda a:a.startswith('s'),names)
list(filtered_names)
```

```python
#map----> modified list, returns new list

#syntax-->map(function, iterable)

names=["priya","sam","vicky","sindhu","krish"]
changed_names=map(str.upper,names)
print(names)
print(list(changed_names))

```

```python
marks=[34,78,90,25,78,90,89,78,90,78,89]

updated_marks=map(lambda x:x+10,marks)

list(updated_marks)
```

```python
#reduce-----> reduce list into single data

#syutax----> reduce(function, iteration)

from functools import reduce
total=[89,90,78,67,56]
final=reduce(lambda x,y:x+y,total)

final
```

```python
# filter the string which are length greater than 5.


s = ["starfish", "koi", "octopus", "ray"]

filtered_s = filter(lambda x: len(x) > 5, s)
print(list(filtered_s)) 
```

```python
#find maximum number in a list

number1=[2,7,9,10,26,5]
max(number1)
```

```python
#find the product of list

marks=(90,80,70,60,50)
from functools import reduce

final=reduce(lambda x,y:x+y,marks)

final
```

```python
# we have list of numbers
# squared only even number

marks = [23, 46, 39, 56, 43, 98, 29, 24, 12, 56, 46]

# Filter even numbers first, then square them

even_numbers = [x for x in marks if x % 2 == 0]
squared_numbers = [x**2 for x in even_numbers]

print(even_numbers)
print(squared_numbers)
```
