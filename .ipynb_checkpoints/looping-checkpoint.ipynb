```python
#Print numbers from 1 to N using a loop

for i in range(0,5):
    print(i)
```

```python
#Print even numbers between 1 to N

for i in range(10,100):
    print (i, end=" ")
```

```python
for i in range(10, 100):
    if i % 2 != 0:
        print(i, end=" ")
```

```python
# Find the sum of digits of a number

N=int(input())
sum=0
for i in range(N+1):
    sum=sum+i
    print(sum)
```

```python
# Print multiplication table of a number 

table_no=int(input("enter the table no"))

for i in range(1,21):
    print(f"{table_no} x {i}= {table_no*i}")

```

```python
# while loop ---- we dont know the number of iteration

while conition:
    block code
    
```

```python
#Sum of N natural numbers (1 + 2 + ... + N)

i=0

while i<=5:
    print(i)
    i=i+1
```

```python
password=input("enter the number")

correct_password="admin@123"

while password!=correct_password:
    print("wrong password, try again")
    password=input("enter the password")

print("Access granted")
```

```python
#Print a right-angled triangle

row=5

for i in range(1,row+1):
    print('*' *i)
```

```python
#looping control statement

break, continue, pass

break----- exits the loop immediately when encountered.
continue---- skip the current iteration.
pass----- it acts as place holder for future code reference
```

```python
for i in range(1,10):
    if i==3:
        break
    print(i)
```

```python
for i in range(1,10):
    if i==3:
        continue
    print(i)
```

```python
#Reverse a number (e.g., 1234 → 4321)

i = input("Enter a number")

reverse = i[::-1]
print(reverse)

```
