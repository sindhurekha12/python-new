```python
def reverse_int():
    
    number = int(input("Enter the number: "))
    rev = 0
    
    while number > 0:
        rem = number % 10
        rev = rem + (rev * 10)
        number = number // 10
    print("Reversed Number:", rev)
reverse_int()
```
