```python
gender = input("Enter gender (male/female): ")
age = int(input("Enter the age: "))

if gender == "male":
    if age < 30:
        print("Married male: insurance amount Rs.10000.")
    elif age < 35:
        print("Unmarried male: insurance amount Rs.5000.")
    else:
        print("Male: Not eligible")

elif gender == "female":
    if age < 28:
        print("Married female: insurance amount Rs.12000.")
    elif age < 32:
        print("Unmarried female: insurance amount Rs.5000.")
    else:
        print("Female: Not eligible")

else:
    print("Invalid gender entered.")
```

```python
# Q.2 ATM machine

name = input("Enter the name: ")
number = int(input("Enter the account number: "))
balance = 50000

if number == 12345:  
    print("1. Withdraw")
    print("2. Balance Inquiry")
    
    choice = int(input("Select an option (1 or 2): ")) 
    if choice == 1:
        amount = int(input("Enter amount to withdraw: "))
        if amount <= balance:
            balance -= amount  # balance = balance - amount
            print(f"Withdrawal successful! Remaining balance: Rs.{balance}")
        else:
            print("Insufficient balance!")
            
    elif choice == 2:
        print(f"Your current balance is: Rs.{balance}")
    else:
        print("Invalid choice.")

else:
    print("Invalid Account Number. Please try again.")

```
