```python
class Laptop:
    def __init__(self,brand,price,ram,rom):   #def __init__ ---- 2 underscore
        self.brand=brand         #self. current object define
        self.price=price
        self.ram=ram
        self.rom=rom
        print("Automatically calls")
    def show(self):
        print("*****************")
        print(f"Laptop brand {self.brand}")
        print(f"Laptop price {self.price}")
        print(f"Laptop ram {self.ram}")
        print(f"Laptop rom {self.rom}")
        print("****************")

L1=Laptop("hp",60000,"64gb",120)
L2=Laptop("lenova",55000,"84gb",130)
L3=Laptop("Dell",90000,"98gb",120)

L1.show()

L2.show()

L3.show()
```

```python
#inheritance ----------> one class inherits properties, methods from another class

#why inhertiance?
code reusability

types of inheritance:
single inheritance
multiple
multilevel 
hierchical
hybrid
```

```python
#single iheritance

class animal:
    def show(self):
        print("Animals.......")

class dog(animal):
    def sound(self):
        print("Dogs makes sound")
        
    def show(self):
        super().show() 
        print("Dogs show")
        animal.show(self)
    
        
d1 = dog()
d1.sound()
d1.show()
```

```python
#multilevel inheritance

class grandparent:
    def work(self):
        print(f"Grandpa reading news paper")

class parent(grandparent):
    def display(self,name):
        print(f"parent name {name}")
    def work(self):
        print(f"parents working in a company")
        super().work()

class child(parent):
    def show(self):
        print(f"child.......")
    def work(self):
        print(f"child studying")
        super().work()

c1=child()
c1.display("priya")
c1.work()
        
```

```python
#multiple

class father:
    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class mother:
    def show(self,name,job):
        print(f"Name:{name}")
        print(f"job:{job}")

class son(father,mother):
    def display(self,name,job):
        print(f"name:{name}")
        print(f"job:{job}")

s1=son()
s1.display("sri","ai engineer")

s1.show("priya","housewife")
```

```python
#multiple

class father:
    def __init__(self,name,job):
        self.name=name
        self.job=job

    def display(self,name,job):
        print(f"Name:{name}")
        print(f"Job:{job}")

class mother:
    def show(self):
        print(f"Name:{self.name}")
        print(f"job:{self.job}")

class son(father,mother):
    def display(self,name,job):
        print(f"name:{name}")
        print(f"job:{job}")
        print("**********")
        super().display("ram","company")

s1=son("guest","Unemployee")
s1.display("sri","ai engineer")

s1.show()
```

```python
#hierchical inheritance

class parent:
    def show(self):
        print("parents......")

class child(parent):
    pass
class son(parent):
    pass

s1=son()
c1=child()

s1.show()
c1.show()
```

```python
# method overriding

import math

class shape:
    def area(self):
        pass

class rectangle(shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def area(self):
        return self.length*self.width


class circle(shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return math.pi*self.radius*self.radius

class triangle(shape):
    def __init__(self,b,h):
        self.b=b
        self.h=h
    def area(self):
        return 0.5*self.b*self.h

shape=[rectangle(10,2),circle(4),triangle(7,6)]

for shape in shape:
    print(shape.area())
        
```

```python
# user define error

class agetoosmall(Exception):
    pass

def checkage(age):
    if age<0:
        raise agetoosmall("Age should be positive")
    else:
        print("Valid age")


try:
    checkage(10)
except Exception as e:
    print(e)
```

```python
class heightcheck(Exception):
    pass

def medium(height):
    if height==5:
        raise heightcheck("medium height")
    else:
        print("Valid height")

def low(height):
    if height==0:
        raise heightcheck("low height")
try:
    medium(10)
except Exception as e:
    print(e)
```
