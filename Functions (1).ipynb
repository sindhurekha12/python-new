```python
#Function without arguments without return value

def welcome(name):  #parameter
    print(f"welcome {name}")

welcome("sri")  #arguments----- actual value
```

```python
# functions arguments type

1.positional arguments
2.keyword arguments
3.default arguments
4.variable length keyword arguments
5.variable length position argumens
```

```python
#positional arguments

def student(name,age,course):
    print("name",name)
    print("age",age)
    print("course",course)

student(24,"sri","ai")
```

```python
#default arguments

def data(name="guset"):
    print(f"welcome {name}")

data()
data("priya")
```

```python
# Keyword arguments

def details(name,city,age,interest):
    print(name)
    print(city)
    print(age)
    print(interest)
details(name="vishnu",age=27,city="cbe",interest="cricket")
```

```python
#varaible length positional arguments

def total(*args):
    return sum(args)

total(90,89,78,98)
```

```python
#varaible length keyword arguments

def data(**kwargs):
    for key,value in kwargs.items():
        print(key,value)


data(name="priya",age=24,city="cbe",country="india",interest="ancing",hobby="cooking")
```

```python
# Variable
# local variable

x=20
def my_func():
    x = 10 # local variable
    print(x)

my_func()
```

```python
# global variable
x=20
def my_func():
    global x   # used to change outer function
    print(x)

my_func()
print(x)
```

```python
# non local variable

def outer():
    x = "outer value"

    def inner():
        nonlocal x   # used to change the inner function
        x = "change in inner"

    inner()
    print(x)

outer()
        
```
