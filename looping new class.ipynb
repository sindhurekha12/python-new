```python
number=int(input("Enter the number:"))
rev=0
while number>0:
    rem=number%10
    rev=rem+(rev*10)
    number=number//10
print("reversed number:",rev)
```

```python
number=int(input("Enter the number: "))
temp=number
count=0
while temp>0:
    count=count+1
    temp=temp//10
temp=number    # restore value
add=0       # value add 
while temp>0:
    rem=temp%10
    add=add+pow(rem,count)
    temp=temp//10
if number==add:
    print(f"{number} is an amstrong no!!!...")
else:
    print(f"{number} is not an amstrong no!!!..")
```
