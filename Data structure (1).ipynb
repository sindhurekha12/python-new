```python
#operators

#arithmetic operators
#assignment operators
#comparison or relational operators
#logical operators
#bitwise operators
#identity operators
#membership operators
```

```python
number1=int(input())
number2=int(input())

print(number1+number2)
print(number1-number2)
print(number1*number2)
print(number1**number2)   #a**b is equivalent to a power b # 10*10*10*10*10 = 100000
print(number1/number2)
print(number1//number2)  #10 divided by 5 is 2.00. Floor division chops off the decimal, leaving just 2.
print(number1%number2)   # % is to return the remainder of a division between two numbers
```

```python
#relational or comparison operators

number1=int(input())
number2=int(input())

print(number1>number2)
print(number1<number2)
print(number1>=number2)
print(number1<=number2)
print(number1==number2)
print(number1!=number2)
```

```python
#assignment operators

a=10
a+=12  #a=a+12

print(a)

b=12
b*=a   #b=b*a
print(b)
```

```python
#logical operators

a=10
b=12

print(a>b and a!=b)
print(a!=b or a==b)
print(not a==b)
```

```python
#bitwise operators

a=14
b=14

print(a&b)
print(a|b)
print(a^b)
print(~a)   #-(a+1)
```

```python
#membership operator------>in, not in

a=("apple","banana","cherry")
print("apple" in a)
print("guva" not in a)
```

```python
#identity operator----------->is, isnot


b=a
c="priya"
print(a is b)
print(a is not c)
```
